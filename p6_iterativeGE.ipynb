### 1. **p6.m** - low-rank approximation of an image by Gaussian elimination with complete pivoting
#### This code takes an image, converts it to grayscale, and iteratively approximates it using rank-1 matrix updates. Every k = 600 iterations for `image.jpg` and k = 75 iterations for `image2.jpg`, the current approximation is displayed along with the rank and root mean square error. The loop continues until the approximation reaches the full rank of the image, with the final approximation displayed at the end. The code is designed to give visual feedback during execution. It will update you on the approximation process.

- you will get `RuntimeWarning` after these complete because the `total_rank % chosen_num != 0` and we are displaying the plots on every iteration where `iterative_rank % chosen_num = 0`. This means you won't get the final image, only close to it. This is just my example so that you can see the steps without having to wait too long. Change it as you like. I chose the `chosen_num` based on the size of each image's rank and how long I felt like waiting...

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
from numpy.linalg import norm
```

```python
## this will load an image and convert to grayscale and flt.pt.
A = io.imread('image.jpg')       
A = color.rgb2gray(A)              
A = np.array(A, dtype=float) / 255  

## initialize zeros and rank
Aapprox = np.zeros_like(A)
rankA = np.linalg.matrix_rank(A)   
sq = np.sqrt(A.size)                

## being the iterative approximation of A
for k in range(rankA + 1):
    maxerr = np.max(A)              
    if k % 600 == 0: ## 600 is the chosen_num here. this should print out a plot less than 6 times. 
        ## I chose this "600" value so you won't have to wait too long to get the idea. 
        err = norm(A, 'fro') / sq
        plt.imshow(Aapprox, cmap='gray')   ## displays the current rank approximation as an image
        plt.title(f'rank = {k},  rmserr = {err:.4f}')
        plt.draw()
        plt.pause(0.1) ## be patient as this generates several plots one after the other
        print("Be patient, I am working...")
    
    i, j = np.unravel_index(np.argmax(A, axis=None), A.shape)  
    Ak = np.outer(A[:, j], A[i, :] / maxerr)  
    Aapprox = Aapprox + Ak                    
    A = A - Ak                                
    
plt.show();  
```
